```python
import json
import requests as req
from bs4 import BeautifulSoup
```

```python
big_data={}
for i in range(1,145):
    s="https://en.wikipedia.org/wiki/Ward_No._"+str(i)+",_Kolkata_Municipal_Corporation"
    res=req.get(s)
    soup=BeautifulSoup(res.text,'html.parser')
    soup.prettify()
    lat_long=soup.find_all('span',class_="geo-dec")
    s=(lat_long[0].text)
    #print(s)
    ll=s.split()
    
    #print ( ll )
    pin_='NA'
    try:
        
        pin=soup.find_all('span',class_="postal-code")
        pin_=pin[0].text
    except:
        pin="NA"
    #print(pin[0].text)
    
    data_= soup.find_all('tr',class_="mergedrow")
    l=[]
    data={}
    for j in data_:
        #print('%')
        #print((j))
#         print('\n\n\n')
        d=(j.find_all('td'))
        #print (d.text)
        #print(d)
        try:

            
            a1=((d[0].find('a')).text)
            f=1
            try:
                
                a=(d[0].text)
                if (str(a1)!=str(a)):
                    
                    l.append(str(a1)+" "+str(a))
                else:
                    l.append(str(a1))
            except:
                l.append(a1)
        except:
            
            l.append((d[0].text))
            print("")
        
        
    k=0   
    #print (len(l))
    if (len(l)==8):
        k=0
    else :
        k=1
    
    print(i,'#########')
    print(l)
    temp3={'Ward_no':i}
    temp={'Neigbourhood':l[2]}
    temp1={'PC':l[4-k]}
    temp2={'AC':l[5-k]}
    temp4={'Latitude':ll[1]}
    temp5={'Longitude':ll[0]}
    temp6={'PIN':pin_}
    
    
    
    data.update(temp3)
    data.update(temp)
    data.update(temp1)
    data.update(temp2)
    data.update(temp4)
    data.update(temp5)
    data.update(temp6)
    
    print(data)
    
    big_data.update({i:data})
    
    
#     if i==7:
        
#         break
    
```

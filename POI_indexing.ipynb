Calculating indexing for all POIs storewise

```python
big_data={}
f=0
for i in db.comp.find({}):
    #print (i)
    p=db.coll.find({ "geometry": { '$near': { '$geometry': i['geometry'] ,'$maxDistance' : 3 } } })
    
    data={}
    data_={}
    for k in p:
        poi_data=k['data']
        for j in poi_data:
            
            
            if j in data:
                
                try:
                    
                    data[j]+=(int)(poi_data[j])
                except:
                    if ('.' in poi_data[j]):
                        s=(poi_data[j])[0:len(poi_data[j])-2]
                        data[j]+=(int)(s)
                    
                    
            else:
                
                try:
                    temp={j:(int)(poi_data[j])}
                    data.update(temp)
                except:
                    
                    if ('.' in poi_data[j]):
                        s=(poi_data[j])[0:len(poi_data[j])-2]
                        temp={j:(int)(s)}
                        data.update(temp)
    
    data_.update({'id':i['id']})
    data_.update({'data':data})
    temp_={i['id']:data_} 
    big_data.update(temp_)
    data={}
    
   
    #break
#     if (f==3):
#         break
#     f=f+1
    
print (big_data)
```
